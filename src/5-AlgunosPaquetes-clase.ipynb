## StatsPlots

Es un paquete que agrega funciones de graficacion tipicas del area de estadistica. ] add StatsPlots

```julia
using StatsPlots
using DataFrames
```

```julia
# Generar datos de las mediciones
M = 100  # Cantidad de mediciones
N = 5    # Cantidad de veces que repito el experimento
datos = randn(M,N)
ms = 1:M
```

```julia
plot(ms, datos)
```

```julia
boxplot(ms, datos)
```

```julia
scatter(ms, datos[:,1], markersize=10*datos[:,1])
```

```julia
ms
```

```julia
datos
```

```julia
datos[:,1]   # [indicesRenglon , indicesColumna]
```

```julia
p1 = histogram(datos[:,1], bins = 20)
p2 = histogram(datos[:,2], bins = 20)
p3 = histogram(datos[:,3], bins = 20)
p4 = histogram(datos[:,4], bins = 20)

plot(p1, p2, p3, p4)
```

```julia
l = @layout [ [a; b] c ]  # a,b,c son variables sin significado
plot(p1, p2, p3, layout=l)
```

```julia
errorline(ms, datos)
```

## Usando DataFrames

```julia
df_tec21 = DataFrame(a = -5:4, b = rand(10), c = 10*rand(10))
```

```julia
df_tec21[:, :b]
```

```julia
df_tec21.b
```

```julia
@df df_tec21 plot(:a, :b)
@df df_tec21 plot!(:a, :c)
```

```julia
@df df_tec21 plot(:a, [:b :c])
```

```julia
@df df_tec21 scatter(:a, :b, markersize=10*:c)
```

```julia
@df df_tec21 boxplot([:a :b :c])
```

```julia
@df df_tec21 errorline(:a, [:b :c])
```

## Comentarios de despacho multiple

```julia
function fun(z)
    y = z+1
    return y
end
```

```julia
cc = 1.0
fun(cc)
```

```julia
function fun2(z::Float64)
    y = z+1
    return y
end
```

```julia
function fun2(z::Int64)
    y = z+1
    return y
end
```

```julia
fun2(1)
```

```julia
function fun3(z::T) where T <: Union{Float64, Int64}
    y = z+1
    return y
end
```

```julia
fun3(1)
```

```julia
fun3(1.0)
```

```julia
@code_warntype fun2(1)
```

```julia
@code_warntype fun2(1.0)
```

```julia
@code_warntype fun3(1)
```
